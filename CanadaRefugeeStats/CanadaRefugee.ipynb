```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('can-stats-2012-22.csv')
```

```python
# Check missing values
pd.DataFrame(data.isnull().sum(), columns=['Missing Values'])
```

```python
# Question 1: Countries From Which Canada Has Admitted Highest Number of Refugees

# Create a DataFrame with the top countries and their total refugees sum, then sort values in desc order
# top_unhcr_refugees = data.groupby('Country-of-origin')['UNHCR-refugees'].sum().reset_index(name='UNHCR-refugees').sort_values(by='UNHCR-refugees', ascending=False).head(10)

top_total_refugees = data.groupby('Country-of-origin')['total-count'].sum().reset_index(name='Total refugees').sort_values(by='Total refugees', ascending=False).head(50)

group_2_countries = data.groupby('Country-of-origin')['total-count'].sum().reset_index(name='Total refugees').sort_values(by='Total refugees', ascending=False)[50:100]
group_3_countries = data.groupby('Country-of-origin')['total-count'].sum().reset_index(name='Total refugees').sort_values(by='Total refugees', ascending=False)[100:150]
group_4_countries = data.groupby('Country-of-origin')['total-count'].sum().reset_index(name='Total refugees').sort_values(by='Total refugees', ascending=False)[150:200]

# Display the DataFrame with the top countries for total Refugees
top_total_refugees.reset_index(drop=True).style.bar()
```

```python
# Create Seaborn bar graph to visualize top 10 countries with highest number of refugees resettled in Canada
plt.figure(figsize=(10,6))
sns.barplot(data=top_total_refugees, x='Country-of-origin', y='Total refugees', palette='mako')

# Add labels and titles
plt.xlabel('Country of Origin')
plt.ylabel('Total refugees')
plt.title('Top Countries of origin for Total Refugees in Canada (2012-2022)')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Display plot
plt.tight_layout()
plt.show()
```
