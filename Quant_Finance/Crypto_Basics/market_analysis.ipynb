```python
import websocket   # pip install websocket-client
import json
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///COINS.db')  # creating sqlite engine with COINS database
```

```python
# Binance endpoint for array of all symbols prices, 
# every time the price changes - the tick comes via web socket:

endpoint = 'wss://stream.binance.com:9443/ws/!miniTicker@arr'
```

```python
def on_message(ws, message):
    # global out
    out = json.loads(message)
    # print(out)
    # importing the "out" data to the DB
    df_import(out)

# creating a websocket, with endpoint and on_message function:
ws = websocket.WebSocketApp(endpoint, on_message=on_message)
ws.run_forever()
```

```python
# function that imports data to the DB

def df_import(data):
    df_ = pd.DataFrame(data)
    df_ = df_[df_['s'].str.endswith('USDT')]  # only USDT pairs
    df_.c = df_.c.astype(float)               # close price to be float
    final = df_[['s','E','c']]                # only those columns from the dataframe
    
    # in the loop insert all to the DB, table name is row.s.values[0] 
    for i in range(len(final)):
        row_ = final[i:i+1]
        row_[['E','c']].to_sql(row_.s.values[0], engine, index=False, if_exists='append')
    
```
