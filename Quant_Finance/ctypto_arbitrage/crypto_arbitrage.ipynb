```python
# function to get prices from the exchange

def getprices(exchange, symbol):
    inst = getattr(ccxt, exchange)()
    df = pd.DataFrame(inst.fetchOHLCV(symbol, limit=60))
    df.columns = ['Time', 'Open', 'High', 'Low', 'Close', 'Volume']  # renaming the columns
    df.set_index('Time', inplace=True)  # setting Time as index and 
    df.index = pd.to_datetime(df.index, unit='ms') # transorm it to datetime format
    df = df.astype(float)  # just in case if some values are in String format
    return df
```

```python
# getting prices from all the exchanges ccxt supports

prices = []
fails = []

for exchange in ccxt.exchanges:
    try:
        prices.append(getprices(exchange, 'BTC/USDT'))
        print('Prices succesfully pulled from ' + exchange)
    except:
        fails.append(exchange)
        print('Failed to pull prices from ' + exchange)
   
```

```python
print('Successfully pulled from ' + str(len(prices)) + ' exchanges.')
print('Failed to pull from ' + str(len(fails)) + ' exchanges.')
```
